```python
import requests
from bs4 import BeautifulSoup


url = "https://atlanticbooks.com/collections/non-fiction"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find all product containers
products = soup.find_all('div', class_='productgrid--items-products-per-row-4')
books = []
for book in soup.find_all('div', class_='productitem--info'):
    title = book.find('h2').text
    author = book.find('p').text


    # Check if the price div exists before accessing its text
    Original_price= book.find('div', class_='price__compare-at visible')
    price_o = Original_price.text if Original_price else 'N/A' # Assign 'N/A' if price is not found

    # Check if the final_price div exists before accessing its text
    final_price = book.find('div', class_='price__current--on-sale')
    price_f = final_price.text if final_price else 'N/A' # Assign 'N/A' if final_price  is not found

    # Check if the availability span exists before accessing its text
    Discount_Status = book.find('span', 'data-price-percent-saved')
    Discount = Discount_Status.text if Discount_Status else 'N/A' # Assign 'N/A' if availability is not found

    books.append({'title': title, 'author': author, 'Original_price': price_o, 'final_price': price_f , 'Discount_Status': Discount }) #'genre': genre,
```

```python
print(Discount)
```

```python
import pandas as pd

df = pd.DataFrame(books)
#Handle potential errors from empty or unexpected values in the 'price' column
if 'price_o' in df.columns:
    # Replace non-numeric values with NaN before converting to float
    df['price_o'] = pd.to_numeric(df['price_o'].str.replace('[$,]', '', regex=True), errors='coerce')
    df['price_f'] = pd.to_numeric(df['price_f'].str.replace('[$,]', '', regex=True), errors='coerce')
# if 'genre' in df.columns:
#     df['genre'] = df['genre'].str.strip().str.lower()
df.drop_duplicates(subset=['title', 'author'], inplace=True)
df.head(30)
```
